# **ASSIGNMENT 2 - DIFFERENT ENSEMBLE LEARNING TECHNIQUES**

```python
import numpy as np
import pandas as pd
```

```python
# READING THE DATA AND FINDING NaN VALUES
data=pd.read_csv('/content/train_loan.csv')
data.head()
```

```python
data.isna().sum()
```

```python
# dropping the variable 'Loan_ID' since it has no significance
data.drop('Loan_ID',axis=1,inplace=True)
```

```python
data.head()
```

```python
#finding the categorical values
data.dtypes
```

```python
# since Dependents contain values with + it is replaced
data['Dependents']=data['Dependents'].str.replace('+','')
data['Dependents'] = pd.to_numeric(data['Dependents'], errors='coerce').fillna(0).astype(int)
```

```python
data.columns
```

```python
# Handling the missing values in the numerical variables-Dependents,LoanAmount,Loan_Amount_Term,Credit_History
data['Dependents'].fillna(data['Dependents'].median(), inplace = True)
data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace = True)
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median(), inplace = True)
data['Credit_History'].fillna(data['Credit_History'].median(), inplace = True)
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.Gender=le.fit_transform(data.Gender)
data.Married=le.fit_transform(data.Married)
data.Education=le.fit_transform(data.Education)
data.Self_Employed=le.fit_transform(data.Self_Employed)
data.Property_Area=le.fit_transform(data.Property_Area)
data.Loan_Status=le.fit_transform(data.Loan_Status)
data.head()
```

```python
data.dtypes
```

```python
data.isna().sum()
```

All the categorical values are encoded and NaN are filled.
Now splitting the target column **"Loan_Status"**

```python
x=data.drop('Loan_Status',axis=1)
y=data['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier(n_estimators=150,random_state=42)
rf_clf.fit(x_train,y_train)
y_pred_rf=rf_clf.predict(x_test)
from sklearn import metrics
print('Accuracy = ',metrics.accuracy_score(y_test,y_pred_rf))
```

**BAGGING CLASSIFIER USING DECISION TREE**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
dct_clf=DecisionTreeClassifier()
bag_clf=BaggingClassifier(base_estimator=dct_clf,n_estimators=100,random_state=42)
bag_clf.fit(x_train,y_train)
y_pred_bag=bag_clf.predict(x_test)
print('Accuracy = ',metrics.accuracy_score(y_test,y_pred_bag))
```

**ADABOOST CLASSIFIER**

```python
from sklearn.ensemble import AdaBoostClassifier
ada_clf=AdaBoostClassifier(n_estimators=150,random_state=42)
ada_clf.fit(x_train,y_train)
y_pred_ada=ada_clf.predict(x_test)
print('Accuracy = ',metrics.accuracy_score(y_test,y_pred_ada))
```

**GRADIENT BOOSTING**

```python
from sklearn.ensemble import GradientBoostingClassifier
grad_clf=GradientBoostingClassifier(n_estimators=40,max_depth=3,random_state=42)
grad_clf.fit(x_train,y_train)
y_pred_grad=grad_clf.predict(x_test)
print('Accuracy = ',metrics.accuracy_score(y_test,y_pred_grad))
```

**XGBOOSTING**

```python
import xgboost as xgb
xgb_clf=xgb.XGBClassifier(n_estimators=40,max_depth=3,random_state=42)
xgb_clf.fit(x_train,y_train)
y_pred_xgb=xgb_clf.predict(x_test)
print('Accuracy = ',metrics.accuracy_score(y_test,y_pred_xgb))
```

**FINAL OUTCOME**
Accuracy of : RandomForestClassifier(77.2%), GradientBoosting(78.9%), AdaBoost(74.8%), BaggingClassifier(76.4%), XGBoosting(76.4%).
**GradientBoosting** has better accuracy.
